```python
from google.cloud import vision
from google.oauth2 import service_account
import matplotlib.pyplot as plt
from IPython.display import Image as IPImage, display
```

```python
def detect_text(image_path):
    # Use your service account JSON key file
    credentials = service_account.Credentials.from_service_account_file('/content/upheld-quanta-422315-v0-f07d6c1d989c.json')
    client = vision.ImageAnnotatorClient(credentials=credentials)

    # Read and display the image
    display(IPImage(filename=image_path))

    with open(image_path, 'rb') as image_file:
        content = image_file.read()
    image = vision.Image(content=content)

    # Perform text detection using the Vision API
    response = client.text_detection(image=image)
    texts = response.text_annotations

    # Check if any text was detected and print it
    if texts:
        print("Detected Text:")
        print(texts[0].description.strip())  # Print only the first detected block of text, neatly
    else:
        print("No text found.")

# Example usage, assuming you have an image at '/content/free.jpg'
detect_text("/content/test1.jpg")
```

```python
def detect_labels(image_path):
    # Use your service account JSON key file
    credentials = service_account.Credentials.from_service_account_file('/content/upheld-quanta-422315-v0-f07d6c1d989c.json')
    client = vision.ImageAnnotatorClient(credentials=credentials)

    # Display the image
    display(IPImage(filename=image_path))

    with open(image_path, 'rb') as image_file:
        content = image_file.read()
    image = vision.Image(content=content)

    # Perform label detection using the Vision API
    response = client.label_detection(image=image)

    # Display the full API response
    print("GCP Answer:")

    # Print detected labels separately
    print("Detected Labels:")
    if response.label_annotations:
        for label in response.label_annotations:
            print(label.description)
    else:
        print("No labels found.")

# Example usage, assuming you have an appropriate image for label detection
detect_labels("/content/test1.jpg")
```
